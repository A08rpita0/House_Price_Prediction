```python
import os
import pytesseract
from PIL import Image
import pandas as pd
from tqdm import tqdm

# If you're using Colab, no need to set tesseract_cmd manually
# (it's pre-installed once you run: !apt-get install tesseract-ocr -y)

def extract_text_from_folder(base_dir):
    data = []
    for label_folder in os.listdir(base_dir):
        folder_path = os.path.join(base_dir, label_folder)
        if not os.path.isdir(folder_path):
            continue

        for img_file in tqdm(os.listdir(folder_path), desc=label_folder):
            if img_file.lower().endswith(('.png', '.jpg', '.jpeg')):
                img_path = os.path.join(folder_path, img_file)
                try:
                    text = pytesseract.image_to_string(Image.open(img_path))
                    data.append({
                        'folder': label_folder,
                        'image': img_file,
                        'text': text
                    })
                except Exception as e:
                    print(f"‚ùå Error reading {img_file}: {e}")
    return pd.DataFrame(data)

# üß† Update the paths to match your dataset structure
train_path = "Gambo/Train"
test_path  = "Gambo/Test"

train_df = extract_text_from_folder(train_path)
test_df  = extract_text_from_folder(test_path)

# üíæ Save extracted text to CSV files
train_df.to_csv("train_text.csv", index=False)
test_df.to_csv("test_text.csv", index=False)

# üëÄ Show the first few rows to confirm extraction
train_df.head()

```
